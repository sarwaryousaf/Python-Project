```python
def sumList(lst):
    s=0
    for num in lst:
        s+=num
    print(s)
sumList([2,3,4,5,6])
```

```python
def max_num(lst):
    max_num=lst[0]
    if len(lst)!=0:
        
        for a in lst:
            if a>max_num:
                max_num=a
        print(max_num)
    else:
        print("Empty list")
        
max_num([34,5,66,77,550,43])
```

## default argument

```python
def full_name(first_name,middle_name=" ",last_name=""):
    full_name=f"{first_name} {middle_name} {last_name}"
    print(full_name)
full_name("Muhammad","Yousaf")
```

## Positional Argument

```python
def full_name(first_name,last_name):
    full_name=f"{first_name} {last_name}"
    print(full_name)
full_name("Muhammad","Yousaf")
```

### Keyword Argument

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Data points
x = np.array([0, 1, 2, 3])
y = np.array([0, 5, 0, -5])

# Cubic Spline Interpolation
cs = CubicSpline(x, y, bc_type='natural')

# Generate points for plotting
x_new = np.linspace(0, 3, 100)
y_new = cs(x_new)

# Plot
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, label='Cubic Spline')
plt.legend()
plt.show()
```
